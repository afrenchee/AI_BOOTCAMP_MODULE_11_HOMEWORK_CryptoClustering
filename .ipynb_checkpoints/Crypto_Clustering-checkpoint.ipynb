```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load the data into a Pandas DataFrame and set the index to "coin_id"
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display the first few rows of the original data
market_data_df.head()

# Normalize the data using StandardScaler
scaler = StandardScaler()
market_data_scaled = scaler.fit_transform(market_data_df)

# Create a DataFrame with the scaled data, setting the original index
market_data_scaled_df = pd.DataFrame(market_data_scaled, columns=market_data_df.columns, index=market_data_df.index)

# Display the first five rows of the scaled data
market_data_scaled_df.head()

# Create a list to store inertia values for each k
inertia = []
k_values = range(1, 12)

# Loop through different values of k to compute inertia
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(market_data_scaled_df)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Initialize K-means with the best k value identified (e.g., k=4)
best_k = 4
kmeans = KMeans(n_clusters=best_k, random_state=1)
kmeans.fit(market_data_scaled_df)

# Predict the clusters
clusters = kmeans.predict(market_data_scaled_df)

# Add the cluster values to a new column in the original DataFrame
clustered_df = market_data_df.copy()
clustered_df['Cluster'] = clusters

# Scatter plot with clusters
clustered_df.plot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", c="Cluster", colormap="viridis", colorbar=True)
plt.title('Cryptocurrency Clusters by 24h and 7d Price Changes')
plt.show()

# Reduce to 3 components
pca = PCA(n_components=3)
market_data_pca = pca.fit_transform(market_data_scaled_df)

# Get the explained variance
explained_variance = pca.explained_variance_ratio_.sum()
print(f'Total explained variance by 3 principal components: {explained_variance:.2f}')

# Convert PCA data to DataFrame
market_data_pca_df = pd.DataFrame(market_data_pca, columns=['PC1', 'PC2', 'PC3'], index=market_data_scaled_df.index)

# Display the first five rows of the PCA DataFrame
market_data_pca_df.head()

# Repeat the elbow method on PCA-reduced data
pca_inertia = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(market_data_pca_df)
    pca_inertia.append(kmeans.inertia_)

# Plot elbow curve for PCA data
plt.figure(figsize=(10, 6))
plt.plot(k_values, pca_inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k (PCA Data)')
plt.show()

# Initialize K-means with the best k for PCA data
best_k_pca = 4
kmeans_pca = KMeans(n_clusters=best_k_pca, random_state=1)
kmeans_pca.fit(market_data_pca_df)

# Predict the clusters
pca_clusters = kmeans_pca.predict(market_data_pca_df)

# Add clusters to PCA DataFrame
market_data_pca_df['Cluster'] = pca_clusters

# Scatter plot for PCA clusters
market_data_pca_df.plot.scatter(x='PC1', y='PC2', c='Cluster', colormap='viridis', colorbar=True)
plt.title('Cryptocurrency Clusters by PCA Components')
plt.show()

# Get the weights of each feature on each principal component
pca_weights = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3'], index=market_data_scaled_df.columns)

# Display the feature weights
pca_weights

# Export final clustered data to CSV without the index
clustered_df.to_csv("crypto_clustered_data.csv", index=False)
```
